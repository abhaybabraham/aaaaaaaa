```
season = input("Enter the season : Summer,Winter or Rainy")
season = season.lower()
if season == 'summer':
  print("It is Hot")
elif season == 'winter':
  print('It is cold')
elif season == 'rainy':
  print('It is Monsoon time')
else:
  print('Wrong input')
```
